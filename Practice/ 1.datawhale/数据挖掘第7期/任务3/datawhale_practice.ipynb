### 使用IV值筛选特征

```python
def cal_WOE(dataset, col, target):
    # 对特征进行统计分组
    subdata = pd.DataFrame(dataset.groupby(col)[col].count())
    # 每个分组中响应客户的数量
    suby = pd.DataFrame(dataset.groupby(col)[target].sum())
    # subdata 与 suby 的拼接
    data = pd.DataFrame(pd.merge(subdata, suby, how='left', left_index=True, right_index=True))

    # 相关统计，总共的样本数量total，响应客户总数b_total，未响应客户数量g_total
    b_total = data[target].sum()
    total = data[col].sum()
    g_total = total - b_total

    # WOE公式
    data["bad"] = data.apply(lambda x:round(x[target]/b_total, 100), axis=1)
    data["good"] = data.apply(lambda x:round((x[col] - x[target])/g_total, 100), axis=1)
    data["WOE"] = data.apply(lambda x:np.log(x.bad / x.good), axis=1)
    return data.loc[:, ["bad", "good", "WOE"]]


def cal_IV(dataset):
    dataset["IV"] = dataset.apply(lambda x:(x["bad"] - x["good"]) * x["WOE"], axis=1)
    IV = sum(dataset["IV"])
    return IV
```

```python
data_IV = pd.DataFrame()
fea_iv = []

col_list = [col for col in x.drop(labels='status', axis=1)]

for col in col_list:
        col_WOE = cal_WOE(x, col, 'status')
        col_IV = cal_IV(col_WOE)
        if col_IV > 0.1:
            data_IV[col] = [col_IV]
            fea_iv.append(col)

fea_iv
```

```python
univ = [col for col in x.columns if col not in fea_iv]
univ
```

### 训练集测试集划分

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=2018)
```

```python
x_train.head()
```

### lr

```python
lr = LogisticRegression(penalty='l1', class_weight='balanced', max_iter=100, random_state=2018)
lr.fit(X=x_train, y=y_train)
y_valid = lr.predict_proba(x_test)[:, 0]

roc_score = roc_auc_score(y_test, y_valid)
print(roc_score)
```

### SVM

```python
svc = SVC(class_weight='balanced', probability=True, random_state=2018)
svc.fit(X=x_train, y=y_train)

y_valid = svc.predict_proba(x_test)[:, 0]

roc_score = roc_auc_score(y_test, y_valid)
print(roc_score)
```

### Tree

```python
dtree = DecisionTreeClassifier(max_depth=30, min_samples_split=3, max_features='log2', random_state=2018, \
                               max_leaf_nodes=8, class_weight='balanced')
dtree.fit(X=x_train, y=y_train)

y_valid = dtree.predict_proba(X=x_test)[:, 0]

roc_score = roc_auc_score(y_test, y_valid)
print(roc_score)
```

### rf

```python
rf = RandomForestClassifier(class_weight='balanced', random_state=2018)
rf.fit(X=x_train, y=y_train)
y_valid = rf.predict_proba(X=x_test)[:, 0]
roc_score = roc_auc_score(y_test, y_valid)
print(roc_score)
```
